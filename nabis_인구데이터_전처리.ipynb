```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/nabis/세종시_연령인구.csv")
```

##동별 어린이 인구

```python
child_df=df.drop(columns=['연령','Unnamed: 1','세종특별자치시'])

child_df=child_df.iloc[[3*i for i in range(1,20)]]
child_df.index=[f'{i}세' for i in range(19)]
child_df
```

```python
# 데이터프레임 전체를 정수형으로 변환
child_df = child_df.apply(pd.to_numeric, errors='coerce').astype(int)
child_df.info()
```

```python
summed_df = pd.DataFrame(child_df.sum(axis=0))
summed_df.columns=['어린이인구수']
summed_df
```

##동별 출산율

조출생률 (Crude Birth Rate, CBR): \
(조출생률은 특정 연도 내의 총 출생아 수 / 해당 연도의 전체 연앙인구(1년간의 평균 인구))*1,000 \
- 총 출생아 수는 동별 0세 인구로 정의 \
- 연앙인구는 동별 전체 인구 수를 합산


조출생률의 범위 \
- 낮은 조출생률: 10 이하 (1,000명당 10명 미만의 출생)
- 중간 조출생률: 10`20 (1,000명당 1020명 사이의 출생)
- 높은 조출생률: 20 이상 (1,000명당 20명 이상의 출생)


```python
birth_df=df.drop(columns=['Unnamed: 1','세종특별자치시'])
birth_df=birth_df.iloc[[0,3]]
birth_df=birth_df.drop(columns=['연령'])
birth_df.index=['합계','0세']
district=birth_df.columns
birth_df
```

```python
baby=birth_df.iloc[1].astype(int).tolist()
total=birth_df.iloc[0].str.replace(',','').astype(int).tolist()
birth_df=pd.DataFrame(np.array(baby)/np.array(total)*1000)
birth_df.columns=['조출생율']
birth_df.index=district
birth_df
```

##어린이 인구수와 조출생율 데이터프레임 병합

```python
final=pd.concat([summed_df,birth_df],axis=1)
final
```

```python
final.to_csv('final.csv')
```
